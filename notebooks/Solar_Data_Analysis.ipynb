```python
import pandas as pd
data1 = pd.read_csv('../data/benin-malanville.csv')
data2 = pd.read_csv('../data/togo-dapaong_qc.csv')
data3 = pd.read_csv('../data/sierraleone-bumbuna.csv')

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

print(f'THIS IS THE FIRST DATA SET: \n {df1.head()}')
print(f'THIS IS THE FIRST DATA SET: \n {df2.head()}')
print(f'THIS IS THE FIRST DATA SET: \n {df3.head()}')

```

```python
missed1 = df1.isnull().sum()
missed2 = df2.isnull().sum()
missed3 = df3.isnull().sum()
print(f'This is the missed values: \n{missed1}')
print(f'This is the missed values: \n{missed2}')
print(f'This is the missed values: \n{missed3}')
```

```python
df1_duplicated = df1[df1.duplicated]
df2_duplicated = df2[df2.duplicated]
df3_duplicated = df3[df3.duplicated]
print(len(df1_duplicated))
print(len(df2_duplicated))
print(len(df3_duplicated))
```

```python
df1['Comments'] = df1['Comments'].fillna(0)
df2['Comments'] = df2['Comments'].fillna(0)
df3['Comments'] = df3['Comments'].fillna(0)
print(df1['Comments'].head())
print(df2['Comments'].head())
print(df3['Comments'].head())
```

```python
headers1 = df1.columns
headers2 = df2.columns
headers3 = df3.columns

def mean1():
    for header1 in headers1:
        if pd.api.types.is_numeric_dtype(df1[header1]):
            mean_value = df1[header1].mean()
            print(f'Mean of {header1}: {mean_value}')
        else:
            print(f'{header1} is not a numeric column.')

def mean2():
    for header2 in headers2:
        if pd.api.types.is_numeric_dtype(df2[header2]):
            mean_value = df2[header2].mean()
            print(f'Mean of {header2}: {mean_value}')
        else:
            print(f'{header2} is not a numeric column.')


def mean3():
    for header3 in headers3:
        if pd.api.types.is_numeric_dtype(df3[header3]):
            mean_value = df3[header3].mean()
            print(f'Mean of {header3}: {mean_value}')
        else:
            print(f'{header3} is not a numeric column.')
print(f'{mean1()} {mean2()} {mean3()}')
```
